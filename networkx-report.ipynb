## Jupyter Notebook 출력하기

50개의 노드를 출력하고 100개의 무작위 링크를 생성해보았다.

```python
import networkx as nx
import random

G = nx.Graph()
for n in range(50):
   G.add_node(n)

for i in range(100):
    
   node_a = random.randint(0, 49)
   node_b = random.randint(0, 49)
   G.add_edge(node_a, node_b)

pos = nx.spring_layout(G)
nx.draw_networkx_edges(G, pos, alpha=0.2);
nx.draw_networkx_nodes(G, pos, node_size=50);
```

생성된 네트워크의 현황은 아래와 같다.

```python
print(nx.info(G))
```

```python
import pandas as pd
degree_freq = nx.degree_histogram(G)
df_degree_freq = pd.DataFrame(degree_freq, columns=['Frequency']).rename_axis('Degree')
df_degree_freq.plot.bar();
```

Degree의 분포를 보면 위와 같다. degree가 3-5인 노드들이 64%(32/50)를 차지하고 있다.
